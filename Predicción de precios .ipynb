**Pipeline**\
La funcion Pipeline permite juntar varias transformaciones secuenciales que se requieren ejecutar a un conjunto de datos\

*Sintaxis*
Pipeline(steps = [("TituloTrasnformaicon",Transformacion1),("TituloTrasnformaicon2",Transformacion)....])

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
```

```python
# crear la lista de variables numericas y categoricas
# numericas : 'median_income','total_rooms', 'housing_median_age', 'bedrooms_per_room'
# categoricas : 'ocean_proximity'
numeric_list = ['median_income','total_rooms','housing_median_age','bed_rooms_per_room']
categorical_list = ['ocean_proximity']
```

```python
df.columns
```

```python
# crear el pipeline de procesamiento para datos numericos 
# StandardScaler es para los numeros 
# OneHotEncoder es para las variables categoricas 
numeric_transformer = Pipeline(steps = [('scaler', StandardScaler())])
categorical_transformer = Pipeline(steps =  [('onehot',OneHotEncoder())])
```

<center><font face = 'times new roman' size = 6 color = '#2f3f40'> Columnas Transformer </font></center>

ColumnTransformer permite transformar datos de un dataframe, permite aplicar transformaciones especificas a columnas de un dataframe, las cuales pueden ser numéricas y categóricas 

Para aplicar este metodo se deben aplicar un **Nombre**, **Objeto** y **Columnas**

* Nombre = nombre de la variable (cualquiera)
* Objeto = numeric_transformer y categorical_transformer, definidos en la ceda de arriva 
* Columnas = columnas del dataframe a utilizar

```python
# Definir los transformadores
transformer = ColumnTransformer([
    ('num',numeric_transformer,numeric_list),('cat',categorical_transformer,categorical_list)
])
```

Validacion cruzada
buscar definicion e imagenes

 cross-validation o validación cruzada es un método que permite probar el rendimiento de un modelo predictivo de Machine Learning.
 
 es importante garantizar la exactitud de las predicciones del modelo en producción.
 
 Con el fin de evaluar el rendimiento de un modelo de Machine Learning, hay que probarlo con nuevos datos. En función del rendimiento del modelo con datos desconocidos, se puede determinar si aún falta por ajustarlo, se ha ajustado de más o está “bien generalizado”
 
 Buscar una imagen 

```python
# EVALUACION CRUZADA DE LA REGRECION LINEAL 
# https://scikit-learn.org/stable/modules/model_evaluation.html

from sklearn.model_selection import cross_val_score
linear_score = cross_val_score(model_linear_regression, X,y, scoring= 'neg_root_mean_squared_error', cv = 10)

linear_score
```

```python
# Cambiamos el signo y sacamos la raiz
linear_score_score = np.sqrt((-1)*linear_score)

# eliminamos los nan
outpu1 = linear_score_score[np.isfinite(linear_score_score)]
```

```python
# definimos una funcion que nos devuelva el score, mean y standard deviation

def display_scores(scores):
    print('Score : ', scores)
    print('Mean : ', scores.mean())
    print('Standard Deviation : ', scores.std())
```

```python
display_scores(outpu1)
```

<center><font face = 'times new roman' size = 6 color = '#2f3f40'> Modelo de prediccion : Decision Tree </font></center>

```python
#APLIOCAR EL MODELO DE ARBOLES DE DECISION CON DECISIONTREEREGRESSOR

from sklearn.tree import DecisionTreeRegressor

# 1 Nombrar una variable para nuestro modelo
tree_reg=DecisionTreeRegressor(random_state=42)

# 2 Definir el Pipeline : Transformacion + modelo
modelo_tree_reg = Pipeline(steps= [
    ('transformer',transformer),
    ('tree_reg',tree_reg)
])

# 3 Ajustar el modelo
modelo_tree_reg.fit(X_train, y_train)
```

```python
# evaluacion
# 4.  Calcular el score con la validacion cruzada cv = 10
tree_reg_score = cross_val_score(modelo_tree_reg,X,y,
                                scoring = 'neg_root_mean_squared_error', cv = 10)

# 5. Realizar el cambio de signo y sacar raiz
random_rmse_score = np.sqrt((-1)*tree_reg_score)

# ELIMANDO LOS NAN 
#BUSCA PORQUE NO SE ELIMINARON AL PRINCIPIO 
output2 = random_rmse_score[np.isfinite(random_rmse_score)]

# 6 mostramos los resultados 
display_scores(output2)
```

<center><font face = 'times new roman' size = 6 color = '#2f3f40'> Modelo de prediccion : Random Forest </font></center>

```python
# Aplicar modelo RandomForestRegressor 
from sklearn.ensemble import RandomForestRegressor

# 1. Nombrar una variable para nuestro modelo 
forest_reg = RandomForestRegressor(n_estimators = 10, random_state= 42)
```

```python
output2
```

```python
# 2 definir el pipeline: transformacion + modelo

model_forest_reg = Pipeline(steps = [
    ('transformer',transformer),
    ('forest_reg', forest_reg)
])

# 3 ajustar el modelo 
model_forest_reg.fit(X_train,y_train)

```

```python

# 4 evaluacion del modelo con la vañidacion CV=10
forest_reg_score = cross_val_score(model_forest_reg,X,y,
                                scoring = 'neg_root_mean_squared_error', cv = 10)

# 5 Cambios de signo y sacar la raiz 
forest_rmse_score = np.sqrt((-1)*forest_reg_score)

outpu3 = forest_rmse_score[np.isfinite(forest_rmse_score)]


# 6. Mostrar el resultado usando la funcion display_scores
display_scores(outpu3)
```
